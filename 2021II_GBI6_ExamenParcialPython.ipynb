````python
### Primero se debe cargar todas las librerías necesarias

import pandas as pd # Nos permite abrir el accesion list descargado para el gen HG17
from Bio import Entrez # Permite descarganos las secuencias multiples mediante url. 
from Bio import SeqIO 
from Bio import AlignIO # Para alinear las secuencias
from Bio import Phylo   # Para la construccion del arbol

##################################################################################

# Se lee el archivo descargado mediante pandas, como una tabla
skin_seq = pd.read_table('SPRR4.seq')

# Se realiza un data frame de los primeros 10 ids de la anterior tabla

df = pd.DataFrame(skin_seq[:10])

# Se trasnforma a lista el anterior data frame

first_list = df.values.tolist()
#print(first_list) # Para comprobar que la lista se ha generado con exito, donde se puede visualizar la obtencion
                   # de una multilista

#Se trasnforma la multilista a una lista unica:

ID_list = []
for sublist in first_list:
    for i in sublist:
        ID_list.append(i)

#print(len(ID_list))
#print(ID_list)

# Obtención de las secuencias múltiples
#Ingresar el correo electrónico
Entrez.email = "camyloveomg@gmail.com"

handle=Entrez.efetch(db='nucleotide',  rettype="gb", id=ID_list, retmode="text") # Se descarga la data tipo genbank
                                                                                 # rettype='gb', y se utilizan solo
                                                                                 # los 10 primeros ids

for seq_record in SeqIO.parse(handle, "gb"):
    print("%s %s..." % (seq_record.id, seq_record.description[:50]))
    print("Sequence length %i, %i features, from: %s" % ( len(seq_record), len(seq_record.features), seq_record.annotations["source"], ) )

print('\n De click aquí y descargue el archivo tipo genbank:', handle.url) # Para descargarse la secuencia de tipo genbank (esta debe guardarse en la misma direccion en
                  # donde se encuentran este jupyter notebook)

###############

# Se convierte el archivo descargado a uno tipo fasta

records = SeqIO.parse("sequence.gb", "genbank")
count = SeqIO.write(records, "sequence.fasta", "fasta") 
#print("\nConverted %i records" % count) # El output debe ser= Converted 10 records. Se puede verificar la 
                                        # conversion con print
from Bio.Align.Applications import ClustalwCommandline
import os

# Cargar data multiple y crear alineamientos
### Se emplea el siguiente comando debido a que se esta usando una distribucion de Linux, especificamente
# Kubuntu, al no existir ejecutables en linux, primero se debe asegurar de tener instalado en el 
# ordenador el programa clustalw

from Bio.Align.Applications import ClustalwCommandline
cline = ClustalwCommandline("clustalw", infile="sequence.fasta")
stdout, stderr = cline()
#print(cline)

#### Generacion del arbol

# Leer alineamiento
ClustalAlign = AlignIO.read("sequence.aln", "clustal")
#print(ClustalAlign) # Se puede verificar el alineamiento con print

# Cargar alineamiento 
with open("sequence.aln","r") as aln: 
    alignment = AlignIO.read(aln,"clustal")
print(type(alignment))

# Inicializar la calculadora de Distancias y del modelo 
from Bio.Phylo.TreeConstruction import DistanceCalculator 
calculator = DistanceCalculator('identity') # distancia que hay entre las secuencias mediante el metodo identity

# Se trasnforman las distancias a una matris de distancias
distance_matrix = calculator.get_distance(alignment) # Se extraen las distancias y se las ordena en una matriz
#print(distance_matrix)

# Construccion de arbol basado en la matriz de distancias previamente creada

from Bio.Phylo.TreeConstruction import DistanceTreeConstructor
constructor = DistanceTreeConstructor(calculator)# con esta vareable se creará el arbol finalmente

## Contruccion del arbol:
sequence_tree = constructor.build_tree(alignment)
sequence_tree.rooted = True
#print(sequence_tree)

# Guardar el arbol en formato xml:

Phylo.write(sequence_tree, "sequence_tree.xml", "phyloxml")

# Recuperar el archivo guardado y construir el arbol filogenético

sequence_tree = Phylo.read(file="sequence_tree.xml", format= "phyloxml")

# Arbol elemental en Matplotlib
import matplotlib
import matplotlib.pyplot as plt # creacion de lienso

# Creacion del arbol sobre el lienso
fig = plt.figure(figsize=(40, 40), dpi=100) # create figure & set the size 
matplotlib.rc('font', size=30)              # fontsize of the leaf and node labels 
matplotlib.rc('xtick', labelsize=30)       # fontsize of the tick labels
matplotlib.rc('ytick', labelsize=30)       # fontsize of the tick labels
axes = fig.add_subplot(1, 1, 1)
Phylo.draw(sequence_tree, axes=axes)
fig.savefig("sequence_cladogram")
````

Estos nodos, representan puntos de divergencia.Por otro lado el nodo **inner2**, representa un ancentro en común entre los genes mencionados y XM_036453080.2, que pertenece al genoma de *Pipistrellus kuhlii*.
