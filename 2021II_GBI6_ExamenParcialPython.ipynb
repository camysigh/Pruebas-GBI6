## Ejercicio 4 [2 puntos]

Realice la visualización de los datos del ejercicio 3. 

* ```pubyear```: utilizando el conteo artículos por año, calcule la media y la desviación estandar (asumiendo que la distribución es normal). Realice un gráfico de la distribución normal de sus datos. Coloque el título, leyenda de los ejes "x" e "y", y la leyenda de la curva (utilice las expresiones matemáticas tipos latex de ser necesario).

````python
#Librerias:
import csv
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

### utilizando el conteo artículos por año, calcule la media y la desviación estandar 
##(asumiendo que la distribución es normal). Realice un gráfico de la distribución 
## normal de sus datos. Coloque el título, leyenda de los ejes "x" e "y", y la leyenda de 
## la curva (utilice las expresiones matemáticas tipos latex de ser necesario).

All_Years=pub_year.iloc[:, 1] # Se extrae la primera columna del data frame creado anteriormente

Year_list = All_Years.values.tolist() # Se genera una lista de los datos

#print(Year_list)

year = list(set(Year_list)) # Se desarrolla una lista de elementos únicos

#print(year )

# Para el conteo de articulos por año 
years = []
numb_paper = []

for p in year:
    years.append(p)
    numb_paper.append(Year_list.count(p))

numb_paper.sort

#print (numb_paper)
## Media de los datos:
print ("La media es =",np.mean(numb_paper))

## Desviación estandar de los datos:

print("La desviación estandar es =", np.std(numb_paper))

#Gráfica de distribución normal
print ("\n\nFigura 1: Distribución Normal de PubYear")
mean = np.mean(numb_paper)
stan = np.std(numb_paper)
x = np.linspace(mean - 3*stan, mean + 3*stan, 100)
#Variables de latex
sigma = stan
label = '$\mu = %.1f, \ \sigma^2=%.1f$' %(mean, stan)
y_norm = stats.norm.pdf(x, mean, sigma)
print ("\n\n")
plt.plot(x, y_norm, label=label, c = 'lightseagreen')
plt.xlabel("N° publicaciones")
plt.ylabel("Probabilidad de N° de publicaciones")
plt.title("Distribución Normal")
plt.legend(title = "Parámetros")
plt.show()

######  Realice un boxplot del conteo de autores por artículo. Coloque las leyendas adecuadas 
## para cada eje y el título.

Authors=pub_authors.iloc[:, 1] # Se extrae la primera columna del data frame creado anteriormente

Authors_numb = Authors.values.tolist() # Se genera una lista de los datos

print ("\n\n\tFigura 2: Boxplot de pub_authors")
print ("\n\n")
fig = plt.figure(1, figsize=(9, 6))
ax = fig.add_subplot(111)
bp = ax.boxplot(Authors_numb)
plt.xlabel('Distribución de la cantidad de autores')
plt.ylabel('N° de Autores')
plt.title('Boxplot de autores por artículo contado')
plt.show()

### Realice una nube de puntos (adaptación de su tarea), resalte al menos 3 países con flechas. 
## Coloque las leyendas adecuadas para cada eje y el título.

# Se importa csv y se abre el archivo coordenadas, para de esta manera definir la latitud y la longitud 
# como un par ordenado.

import csv
coordenadas = {}
with open('data/coordenadas.csv') as file:
    csvr = csv.DictReader(file)
    for row in csvr: # Se escoje que etiquetas se usara y se las convierte en filas
        coordenadas[row['name']] = [(row['latitude']),(row['longitude'])]
    
# Se crea las listas para los nombres de los paises, la longitud, latitud, y el conteo de
# repeticiones.

names_countries=coutries.iloc[:, 0] # Se extrae del data frame las columnas de para el nombre de los paises y 
counts_countries=coutries.iloc[:, 1] # para el conteo de los mismos

names_list = names_countries.values.tolist() # Se nombra las listas de cada columna extraida
counts_list = counts_countries.values.tolist() 

pais_name = names_list # se renombran las listas generadas anteriormente
pais_long = [] # se generan listas vacias para la longitud y latitud
pais_lat = []
pais_count = counts_list

# Se populiza las listas antes creadas

for p in pais_name:
    if p in coordenadas.keys():
        pais_lat.append(float(coordenadas[p][0]))
        pais_long.append(float(coordenadas[p][1]))

# Se verifica que las mismas contengan todos los paises, y que a estos les corresponda las coorde-
# nadas y el conteo correspondiente a cada uno. 

#print(pais_name)
#print(pais_lat)
#print(pais_long)
#print(pais_count)

# Se procede a graficar los puntos en un mapa carteciano mediante el paquete matplotlib.pyplot

print('\n\n\tFigura3: Gráfica de nube de puntos')

import matplotlib.pyplot as plt
%matplotlib inline
plt.scatter(pais_long, pais_lat, s = pais_count, c= pais_count)
plt.colorbar()
plt.xlim(-135,170) # Las dimensiones del plano dependen de las listas de coordenadas.
plt.ylim(-45, 70)
# Se añaden algunos paises como referencia y se marcan los que tienen más relevancia.
ard = dict(arrowstyle="->")
plt.annotate('USA', xy = (-95.712891, 37.09024), 
               xytext = (-87.183406, 30.831239), arrowprops = ard)
plt.annotate('Rusia', xy = (105.318756, 61.52401), 
               xytext = (100.318756, 50.016521), arrowprops= ard)
plt.annotate('Brazil', xy = (-51.92528, -14.235004), 
               xytext = (-30.318756, -05.016521), arrowprops= ard)
plt.annotate('Australia', xy = (133.775136, -25.274398), 
               xytext = (120.921327, -10.759275), arrowprops= ard)
plt.annotate('Mexico', xy = (-102.552784, 23.634501), 
               xytext = (-110.552784, 15.634501), arrowprops= ard)
plt.annotate('España', xy = (-3.74922, 40.463667), 
               xytext = (-30.74922, 40.463667), arrowprops= ard)
plt.annotate('China', xy = (104.195397, 35.86166), 
               xytext = (110.510093, 40.339852), arrowprops= ard)
plt.annotate('Noruega', xy = (8.468946, 60.472024), 
               xytext = (11.468946, 64.472024), arrowprops= ard)
plt.annotate('Reino Unido', xy = (-3.435973, 55.378051), 
               xytext = (-20.435973, 65.378051), arrowprops= ard)
plt.annotate('Ecuador', xy = (-78.183406, -1.831239), 
               xytext = (-68.183406, 0.831239), arrowprops= ard)

params = plt.gcf()
plSize = params.get_size_inches()
params.set_size_inches( (plSize[0] * 3, plSize[1] * 3) )
plt.show()
````
